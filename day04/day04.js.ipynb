```javascript
const fs = require('fs');
var data1;
var raw;
```

```javascript
fs.readFile('./data-04.txt','utf8', (err,data)=>{
  if(err){
    console.log(err);
    return
  }
  data1 = raw = data
})
data1
```

```javascript
data1 = data1.split('\r\n').map(x=> x.split(',').map(y=> y.split('-').map(Number)))
data1
```

### Resultado

```javascript
var result = 0

data1.forEach((pair,idx) => {
  const [[x1,y1], [x2,y2]] = pair
  
  if ((x1 <= x2 && y1 >= y2) || (x1 >= x2 && y1 <= y2)){
    result++
  }
});

result
```
