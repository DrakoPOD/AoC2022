```javascript
var fs = require('fs').promises
```

```javascript
var raw

fs.readFile('./data-14.txt','utf8').then(res=>{
  raw = res
})
raw
```

```javascript
data = raw.split('\r\n').map(x=> x.split(' -> ').map(y=>y.split(',').map(Number)))
data
```

```javascript
var maps = {}
var x_max = -Infinity
var x_min= Infinity
var y_max= -Infinity
var y_min= Infinity
data.forEach(p => {
  for(let i = 0; i < p.length - 1; i++){
    let [x,y] = p[i]
    let [x2,y2] = p[i+1]

    let xl = x2 - x
    let yl = y2 - y

    x_max = Math.max(x,x_max)
    y_max= Math.max(y,y_max)
    x_min= Math.min(x,x_min)
    y_min= Math.min(y,y_min)

    while(true){

      while(true){
        let pos = (x + xl) + ',' + (y + yl)
        
        maps[pos] = true

        if (yl == 0) break
        yl >= 0 ? yl-- : yl++
      }
      if (xl == 0) break
      xl >= 0 ? xl-- : xl++
    }
  }
});
```

```javascript
var output = ''
for (let y = 0; y < (y_max + 2); y++){
  for (let x = (x_min - 6); x < (x_max + 3); x++){
    let pos = x + ',' + y
    if(maps[pos]){
      output += '#'
    }else{
      output += '.'
    }
  }
  output += '\n'
}
fs.writeFile('./map.txt',output)
```

```javascript
var sand_pit = {
  x: 500,
  y: 0,

  move: function(infy = false){
    if(maps[this.x + ',' + this.y]){
      return 'o'
    }
    let pos = this.x + ',' + (this.y+1)
    if (maps[pos] == undefined){
      this.y++
      return false
    }
    pos = (this.x-1) + ',' + (this.y+1)
    if(maps[pos] == undefined){
      this.y++
      this.x--
      return false
    }
    pos = (this.x+1) + ',' + (this.y+1)
    if(maps[pos] == undefined){
      this.y++
      this.x++
      return false
    }
    
    pos = this.x + ',' + this.y
    maps[pos] = true
    this.reset()
    return true
  },
  reset: function(){
    this.x = 500
    this.y = 0
  }
}
```

```javascript
var sand = 0
while(true){
  let i = 0
  
  while(true){
    if (i > 100000) break

    if (sand_pit.move()) {
      sand++
      break
    }
    i++
  }
  if (i > 100000) break

}
sand
```
