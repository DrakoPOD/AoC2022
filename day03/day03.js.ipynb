```javascript
const fs = require('fs');
var data1;
var raw;
```

```javascript
fs.readFile('./data-03.txt','utf8', (err,data)=>{
  if(err){
    console.log(err);
    return
  }
  data1 = raw = data
})
data1
```

```javascript
data1 = data1.split('\r\n').map(text=>{
  const half = text.length/2
  
  const bag1 = text.substring(0,half)
  const bag2 = text.slice(half)

  return [bag1, bag2]
})
data1
```

```javascript
let letters = data1.map(([bag1, bag2]) => {
  for(char of bag1) {
    if (bag2.includes(char)){
      return char
    }
  }
})

letters
```

## Resultado

```javascript
letters.map(char => {
  const ascii = char.charCodeAt()

  if (ascii < 91){
    return ascii - 38
  }
  return ascii - 96
}).reduce((num, total)=> num + total, 0)
```

# Parte 2

```javascript
var data2 = []
raw.split('\r\n').forEach((bag,idx)=> {
  const i = Math.floor(idx/3)

  if(!data2[i]){
    data2.push([])
  }
  data2[i].push(bag)
})
data2

```

```javascript
var letters2 = data2.map(([bag1, bag2,bag3]) => {
  for(char of bag1) {
    if (bag2.includes(char) && bag3.includes(char)){
      return char
    }
  }
})

letters2
```

## Resultado

```javascript
letters2.map(char => {
  const ascii = char.charCodeAt()

  if (ascii < 91){
    return ascii - 38
  }
  return ascii - 96
}).reduce((num, total)=> num + total, 0)
```
