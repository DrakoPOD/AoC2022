```octave
const fs = require('fs');
var data1;
```

```octave
fs.readFile('./data1.txt', 'utf8', (err,data)=>{
  if(err){
    console.log(err);
    return
  }
  data1 = data;
})
console.log(data1)
```

```octave
data1 = data1.split('\r\n\r\n').map(x => x.split('\r\n').map(y=> Number(y)))
console.log(data1)
```

```octave
var sumArr = data1.map(x => x.reduce((a,t)=> a + t,0));
console.log(sumArr)
```

```octave
Math.max(...sumArr)
```
